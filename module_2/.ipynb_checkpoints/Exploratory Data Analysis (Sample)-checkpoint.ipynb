# Data preprocessing

## Data types

The dataset conatins the folowing columns:

>* Bar id - identification number of a certain bar
* Company (Maker-if known) — name of the producer
* Specific Bean Origin or Bar Name — where cacao beans come from
* Broad Bean Origin — where cacao beans were grown
* Company Location — where the bar producer is located
* Bean Type — cacao bean type used in bar production
* REF — relevance of a review (the higher the score is, the more relevant the review is)
* Review Date — year of review publication
* Cocoa Percent — cacao percent in a bar
* Rating — bar assessment by a focus-group

* NaN-values
* Duplicate values

```python
def country_check(country):
    lst = []
    for x in cacao.broad_origin:
        if country in str(x).lower() and x not in lst:
            lst.append(x)
            print(x)
```

```python
def fix_countries(x):
    if pd.isnull(x):
        return x
    if x == 'nan':
        return None
    x = x.replace(' ', '')
    x = x.replace('/', ',')
    x = x.replace('&', ',')
    if '(' in x:
        x = x[:x.find('(')]
    
    dr = 'DominicanRepublic'
    x = x.replace('Dom.Rep.', dr)
    x = x.replace('Dom.Rep', dr)
    x = x.replace('DR', dr)
    x = x.replace('DomincanRepublic', dr)
    x = x.replace('DominicanRep.', dr)
    x = x.replace('Domin.Rep', dr)
    x = x.replace('D.R.', dr)
    
    x = x.replace('Gre.', 'Grenada')
    
    x = x.replace('Guat.', 'Guatemala')
    
    if 'Mad' in x and 'Madagascar' not in x:
        x = x.replace('Mad', 'Madagascar')
    x = x.replace('Mad.', 'Madagascar')
    
    v = 'Venezuela'
    if 'Venez' in x and v not in x: 
        x = x.replace('Venez', v)
    elif 'Ven' in x and v not in x:
        x = x.replace('Ven', v)
    else:
        x = x.replace('Ven.', v)
    
    x = x.replace('Ecu.', 'Ecuador')
    
    x = x.replace('Nic.', 'Nicaragua')
    
    x = x.replace('Haw.', 'Hawaii')
    
    x = x.replace('PNG', 'PapuaNewGuinea')
    
    if 'Mex' in x and 'Mexico' not in x:
        x = x.replace('Mex', 'Mexico')
    
    return x
```

```python
cacao.broad_origin = cacao.broad_origin.apply(fix_countries).sort_values()
```

```python
print(f'Number of unique countries: {cacao.broad_origin.nunique()}')
print(f'Number of countries with 10+ frequency: {(cacao.broad_origin.value_counts() > 10).sum()}')
cacao.loc[:, ['broad_origin']].info()
```

## Categorial data

```python
def get_boxplot(column):
    fig, axes = plt.subplots(figsize=(14, 4))
    sns.boxplot(x=column, y='rating',
                data=cacao.loc[cacao.loc[:, column]\
                               .isin(cacao.loc[:, column]\
                               .value_counts().index[:10])])
    plt.xticks(rotation=45, size=15)
    plt.yticks(size=15)
    axes.set_title(f'Boxplot for {column}', size=20)
    plt.show()
```

```python
for column in ['company', 'specific_origin', 'company_location', 'bean_type', 'broad_origin']:
    get_boxplot(column)
```

## Student's t-test

```python
def get_stat_dif(column):
    cols = cacao.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if weightstats.ttest_ind(cacao.loc[cacao.loc[:, column] == comb[0], 'rating'],
                     cacao.loc[cacao.loc[:, column] == comb[1], 'rating'])[1]\
                     <= 0.05 / len(combinations_all):
            print(f'Statistically significant differences detected for a column {column}')
            break
```

```python
for col in ['company', 'specific_origin', 'company_location', 'bean_type', 'broad_origin']:
    get_stat_dif(col)
```
